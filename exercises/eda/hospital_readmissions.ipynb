### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

### B. Provide support for your arguments and your own recommendations with a statistically sound analysis:

### B1. Setup an appropriate hypothesis test. 

In this scenario, an appropriate hypothesis test would involve determining whether the correlation between excess rate of readmissions and number of discharges is statistically significant. Our null hypothesis will be that there is no correlation between the two ($\rho$ = 0), while the alternate hypothesis is that there exists some correlation between the two ($\rho \neq 0$).

### B2. Compute and report the observed significance value (or p-value).

```python
clean_hospital_read_df.head()
```

```python
# Grab relevant data. 
excess_re_ratio = clean_hospital_read_df[clean_hospital_read_df['Excess Readmission Ratio'].notnull()]['Excess Readmission Ratio'].values
n_discharges = clean_hospital_read_df[clean_hospital_read_df['Excess Readmission Ratio'].notnull()]['Number of Discharges'].values
n_hospitals = np.size(n_discharges)
```

```python
# Calculate Pearson correlation coefficient, get p-value. 
rho, p_package = stats.pearsonr(n_discharges, excess_re_ratio)
z = rho * np.sqrt(n_hospitals - 2.0) / np.sqrt(1.0 - rho**2)
p = stats.norm.pdf(z) * 2.0

print('Pearson correlation coefficient: ', rho)
print('p-value (from scipy): ', p_package)
print('p-value ("by hand"): ', p)
```

### B3. Report statistical significance for $\alpha$ = .01.

For $\alpha$ = .01, we may reject the null hypothesis that there is no correlation between excess readmission ratio and the number of discharges.

### B5. Construct another plot that conveys the same information in a more direct manner.

A better option in this scenario would be a "joint plot" (bivariate plot with marginal distributions). In addition to the scatterplot shown above that shows the relationship between excess readmission rate and number of discharges, we get univariate graphs along each axis to shown the distribution of each variable as well. In the context of this problem, this is helpful because it provides us with more information than we would be able to deduce by eye from the scatter plot alone. 

```python
ax = sns.jointplot(n_discharges, excess_re_ratio)
_ = ax.ax_joint.set_xlabel('Number of Discharges')
_ = ax.ax_joint.set_ylabel('Excess Readmission Ratio')
```

And so, we see that while we have a higher number of small hospitals (low number of discharges), the distribution of excess readmission ratios is symmetric (possibly normal) and centered at 1.0. As an added bonus, we also get the correlation coefficient / p-value displayed on the graph for free!
